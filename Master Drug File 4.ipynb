```python
import pandas as pd
```

```python
#create a file that has all NDC-level information for each drug product
#download the 'package' and 'product' files from the FDA website: 
#https://www.fda.gov/drugs/drug-approvals-and-databases/national-drug-code-directory
```

```python
#load the package file

package = pd.read_csv(r'/Users/barrypeterson/IntegraConnect/ndctext/package.txt',sep ='\t', lineterminator ='\r')
```

```python
#this file is a tab-delimited .txt file and will need to be converted to .csv to join with other files
package.to_csv (r'/Users/barrypeterson/IntegraConnect/ndccsv/package.csv', index=None)
```

```python
package.head()
```

```python
#modify the saved csv:
#create the 'NDC11' and 'NDCProduct' columns
#split the 'NDCPACKAGECODE' into three columns that are separated by the "-"
#'NDC11' is created by modifying the appropriate section, then concatenating into 11 digits
#'NDCProduct' is created by removing the trailing 2 digits of 'NDC11'
#format the two new fields as 'text' and save the file under a new name (eg package1)
```

```python
#Jupyter Notebook converts the NDC fields to int64 which removes leading "0's", eliminating its functionality
#prior to uploading the new file: define the data types for specific columns in the package1 file
#by defining 'NDC11' and 'NDCProduct' as 'string' we can retain the leading "0's"

data_types = {
    'PRODUCTID': 'object',  
    'PRODUCTNDC': 'object',   
    'NDCPACKAGECODE': 'object', 
    'NDC11': 'string',
    'NDCProduct': 'string',
    'PACKAGEDESCRIPTION': 'object',
    'STARTMARKETINGDATE': 'Int64',
    'ENDMARKETINGDATE': 'Int64',
    'NDC_EXCLUDE_FLAG': 'object',
    'SAMPLE_PACKAGE': 'object'     
}
```

```python
#import the package1 modified datafile
#ensure 'productid' is the first column to join on later
package1 = pd.read_csv(r'/Users/barrypeterson/IntegraConnect/ndccsv/package1.csv', dtype=data_types)
```

```python
package1.info()
```

```python
package1.head()
```

```python
#can now load the 'product' file to eventually join with the 'package1' file to complete the NDC-level information
```

```python
#load the product file
#the product file is also a tab-delimited .txt file and will need to be converted to .csv
#jupyter notebook requires encoding in 'windows-1252' to load correctly

product = pd.read_csv (r'/Users/barrypeterson/IntegraConnect/ndctext/product.txt', sep='\t',
                  lineterminator='\r',
                       encoding='windows-1252')
```

```python
#convert to .csv to join with other files later
product.to_csv (r'/Users/barrypeterson/IntegraConnect/ndccsv/product.csv', index=None)
```

```python
#the product file contains a number of columns that are unnecessary 
#define the columns needed in the product file and save under a new file name (eg product1)
product1 = product[['PRODUCTID',
                    'PRODUCTTYPENAME',
                    'PROPRIETARYNAME',
                    'PROPRIETARYNAMESUFFIX',
                    'NONPROPRIETARYNAME',
                    'DOSAGEFORMNAME',
                    'ROUTENAME',
                    'LABELERNAME',
                    'SUBSTANCENAME',
                    'ACTIVE_NUMERATOR_STRENGTH',
                    'ACTIVE_INGRED_UNIT',
                    'PHARM_CLASSES',
                    'DEASCHEDULE']]
```

```python
product1.info()
```

```python
product1.head()
```

```python
#the files are both now loaded and ready to join
#each file contains a line-specific identifier in the 'PRODUCTID' field which we can use to join
```

```python
#left join on the ProductID associated with each file 
NDC = package1.merge(product1, on='PRODUCTID', how='left') 
```

```python
NDC.info()
```

```python
NDC.head()
```

```python
#no longer need NDC-join-related information, so will drop the 'ProductID' field

NDC = NDC.drop(['PRODUCTID'], axis=1)
```

```python
#the drug-related NDC file is now created, 
#need to now begin to add-in additional oncology-related information
```

```python
#Download the most recent oncology-NDC information from SEER:
#https://seer.cancer.gov/oncologytoolbox/canmed/ndconc/?ordering=None&&page=490
```

```python
#modify and add the NDC field into an NDC11 field in the initial column area

#define the file columns to upload the NDC11 field:
data_types3 = {
    'NDC11':  'object', 
    'NDC-11 (Package)': 'object',
    'NDC-9 (Product)': 'object',
    'Generic Name': 'object',
    'Brand Name': 'object',
    'Strength': 'object',
    'SEER*Rx Category': 'object',
    'Major Class': 'object',
    'Minor Class': 'object',
    'Administration Route': 'object',
    'Package Effective Date': 'object',
    'Package Discontinuation Date': 'object',
    'Status': 'object'    
}
```

```python
#upload the SEER NDC file

SEER_NDC = pd.read_csv(r'/Users/barrypeterson/IntegraConnect/ndccsv/SEER NDC.csv', dtype = data_types3)
```

```python
SEER_NDC.head()
```

```python
SEER_NDC.info()
```

```python
#join the NDC file to SEER_NDC to obtain baseline oncology information for each NDC
SEER_NDC_File = pd.merge(SEER_NDC, NDC, on='NDC11', how='left')
```
